```python
import json
import os
from collections import defaultdict
```

```python
with open("/home/workboots/Datasets/DHC/variations/new/var_4/targets/with_areas/case_areas.json", 'r') as f:
    case_areas = json.load(f)
```

```python
with open("/home/workboots/Datasets/DHC/variations/new/var_4/targets/with_areas/case_advs.json", 'r') as f:
    case_advs = json.load(f)
```

```python
with open("/home/workboots/Datasets/DHC/variations/new/var_4/targets/with_areas/selected_advs.txt", 'r') as f:
    selected_advs = f.readlines()
selected_advs = list(filter(None, map(lambda x: x.strip("\n"), selected_advs)))
```

```python
train_cases = os.listdir("/home/workboots/Datasets/DHC/variations/new/var_4/cross_val/0_fold/with_areas/train/")
```

```python
train_cases = list(map(lambda x: os.path.splitext(x)[0], train_cases))
```

```python
test_cases = os.listdir("/home/workboots/Datasets/DHC/variations/new/var_4/cross_val/0_fold/with_areas/test/")
```

```python
test_cases = list(map(lambda x: os.path.splitext(x)[0], test_cases))
```

```python
adv_cases = defaultdict(list)
```

```python
for case, advs in case_advs.items():
    for adv in advs:
        if adv in selected_advs:
            adv_cases[adv].append(case)
```

```python
adv_areas = defaultdict(list)
```

```python
for adv, cases in adv_cases.items():
    for case in cases:
        if case in train_cases:
            adv_areas[adv].extend(case_areas.get(case, []))
```

```python
adv_areas = {k: list(set(v)) for k, v in adv_areas.items()}
```

```python
case_adv_sim = defaultdict(list)
```

```python
for case in test_cases:
    advs = case_advs[case]
    case_adv_sim[case] = {adv: len(set(case_areas[case]).intersection(set(adv_areas[adv]))) * 1./len(set(case_areas[case]))
                    for adv in advs if adv in selected_advs}
        
```

```python
avg_sim = [val for cases, advs in case_adv_sim.items()
           for val in advs.values()]
```

```python
sim = sum(avg_sim) * 1./len(avg_sim)
```
