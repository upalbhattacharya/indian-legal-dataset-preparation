## Getting the win-ratio of each advocate for each charge

```python
def avg_win_ratio(p_cases, d_cases, scores):
    case_list = list(p_cases.union(d_cases)) # Is there a need to convert to list to preserve ordering?
    won_ap = np.array([scores[case] for case in case_list]) * np.array([case in p_cases for case in case_list])
    won_pet = np.array([1 - scores[case] for case in case_list]) * np.array([case in d_cases for case in case_list])
    return sum(won_pet + won_ap) * 1./len(case_list) if len(case_list) != 0 else 0
```

```python
charge_adv_win_ratios = defaultdict(lambda: dict())
for charge in charges:
    for adv in advs.values():
        case_list = set(adv_cases[adv]).intersection(set(charge_cases[charge]))
        p_cases = case_list.intersection(set(p_d_info[adv]["Petitioner"]))
        d_cases = case_list.intersection(set(p_d_info[adv]["Defendant"]))
        charge_adv_win_ratios[charge][adv] = avg_win_ratio(p_cases,d_cases, decision)
        print(len(case_list), len(p_cases), len(d_cases))
    # Order by score
    charge_adv_win_ratios[charge] = {
                                    k: v
                                    for k, v in sorted(
                                        charge_adv_win_ratios[charge].items(),
                                        key=lambda x: x[1], reverse=True)}
```
