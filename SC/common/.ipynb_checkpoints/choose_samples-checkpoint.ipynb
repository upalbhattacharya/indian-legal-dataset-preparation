```python
import json
from collections import Counter
```

```python
with open("case_statute_info.json", 'r') as f:
    case_statute_info = json.load(f)
```

```python
with open("section_freq.json",'r') as f:
    section_freq = json.load(f)
```

```python
with open("act_freq.json", 'r') as f:
    act_freq = json.load(f)
```

## Acts

```python
act_tail = 10
```

#### Number of Acts with less than `act_tail` cases.

```python
tail_acts = [act for act, val in act_freq.items() if val <= act_tail]
```

```python
len(tail_acts)
```

```python
tail_acts
```

#### Number of cases citing acts from the tail

```python
act_tail_cases = [case for case, statutes in case_statute_info.items()
                  if set(statutes["acts"]).intersection(set(tail_acts)) != set()]
```

```python
len(act_tail_cases)
```

#### Number of cases citing only acts from the tail

```python
act_tail_only_cases = [case for case, statutes in case_statute_info.items()
                  if set(statutes["acts"]) < set(tail_acts) and set(statutes["acts"]) != set()]
```

```python
len(act_tail_only_cases)
```

### Acts where each section is cited at least 11 times

```python
main_secs = set(section_freq.keys()) - set(tail_secs)
```

```python
len(main_secs)
```

```python
main_acts = Counter(list(map(lambda x: x.split("_")[0], main_secs)))
```
