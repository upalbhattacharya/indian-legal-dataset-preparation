# Select only correct portions of the text

```python
from bs4 import BeautifulSoup as bs
import os
```

```python
path = "/home/workboots/Datasets/DHC/raw/html_doc/144689556.html"
```

```python
with open(path, 'r') as f:
    raw = f.read()
```

```python
soup = bs(raw)
```

```python
parts = soup.findAll('div', class_=["docsource_main", "doc_title", "doc_author"])
for part in parts:
    part.decompose()
parts = soup.findAll('pre')
for part in parts:
    part.decompose()
```
