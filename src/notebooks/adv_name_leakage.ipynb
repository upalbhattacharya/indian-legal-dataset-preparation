# Checking leakage of advocate names in documents

```python
import os
import json
import re
from tqdm import tqdm
from collections import defaultdict
```

```python
path = "/home/workboots/Datasets/DHC/common/preprocess/fact_sentences/"
```

```python
case_advs_path = "/home/workboots/Datasets/DHC/common/targets/case_advs.json"
```

```python
with open(case_advs_path, 'r') as f:
    case_advs = json.load(f)
```

```python
names = case_advs['70375874']
```

```python
names = [re.findall(r"[A-Z][^A-Z]*", name) for name in names]
```

```python
names
```

```python
names = [r".?\s*".join(name) for name in names]
```

```python
names
```

```python
leaks = {}
```

```python
for case, advs in tqdm(case_advs.items()):
    adv_regex = re.compile(r"|".join([r".?\s*".join(re.findall(r"[A-Z][^A-Z]*", name)) for name in advs]))
    if not os.path.exists(os.path.join(path, f"{case}.txt")):
        continue
    with open(os.path.join(path, f"{case}.txt"), 'r') as f:
        text = f.read()
    leakage = adv_regex.findall(text)
    if leakage != []:
        leaks[case] = leakage
```

```python
len(leaks)
```

```python
with open("/home/workboots/Project_Files/advocate_recommendation/leaked_cases.json", 'w') as f:
    json.dump(leaks, f, indent=4)
```

```python
adv_nums_path = "/home/workboots/Datasets/DHC/common/adv_info/adv_cases_num.json"
```

```python
with open(adv_nums_path, 'r') as f:
    adv_nums = json.load(f)
```

```python
leaked_advs = defaultdict(list)
```

```python
for case, advs in leaks.items():
    names = set([re.sub("\s|\.", "", name) for name in advs])
    for name in names:
        print(name)
        if adv_nums.get(name, -1) == -1:
            continue
        print(adv_nums[name])
        leaked_advs[name].append(case)
    
```

```python
len(leaked_advs)
```

```python
with open("/home/workboots/Project_Files/advocate_recommendation/leaked_adv_cases.json", 'w') as f:
    json.dump(leaked_advs,f, indent=4)
```

```python
leaked_nums = {
    k: len(v) for k, v in sorted(leaked_advs.items(), key=lambda x: len(x[1]), reverse=True)}
```
