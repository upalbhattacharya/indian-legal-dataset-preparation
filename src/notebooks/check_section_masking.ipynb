# Masking Sections

```python
import json
import os
import re
```

```python
path = "/home/workboots/Datasets/DHC/common/preprocess/sentence_sep_json_no_roles/20730536.json"
case_section_info_path = "/home/workboots/Datasets/DHC/common/act_chapter_section_info/case_act_chapter_section_info.json"
```

```python
with open(path, 'r') as f:
    data = json.load(f)
```

```python
with open(case_section_info_path, 'r') as f:
    case_sections = json.load(f)
```

```python
alt_case_section_regex = re.compile("|".join(["Section\s+{}\s+of\s+the\s+Code".format(sec.split("_")[-1]) for sec in case_sections["20730536"]["sections"]]))
```

```python
case_section_regex = re.compile("|".join([sec for sec in case_sections["20730536"]["sections"]]))
```

```python
for idx, content in data.items():
    replace_texts = []
    if content["sections"]:
        print(content["text"])
        for sec, spans in content["sections"].items():
            replace_texts.extend([content["text"][start:end] for (start, end) in spans])
        content["text"] = re.sub("|".join(replace_texts), "[SECTION]", content["text"])
        content["text"] = case_section_regex.sub("[SECTION]", content["text"])
        content["text"] = alt_case_section_regex.sub("[SECTION]", content["text"])
        print(content["text"])
    del content["span"]
    content["sections"] = list(content["sections"].keys())
```
